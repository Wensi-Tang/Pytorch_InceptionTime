```python
from utils.dataloader.TSC_data_loader import TSC_data_loader, UCI_data_loader, MITdb_data_loader
import os
from os.path import dirname
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


device = "cuda:0"

# example_data_set_list = ['BME','CBF','ECG200','FiftyWords', 'ShapeletSim']

dataset_name = 'FiftyWords'
dataset_path = dirname("./Example_dataset/UCRArchive_2018/")

X_train, y_train, X_test, y_test = TSC_data_loader(dataset_path, dataset_name)
```

```python
from Classifier.Easy_use_trainner import Easy_use_trainner
from Classifier.InceptionNet import InceptionNet

Result_log_path = './Example_Results/Results_of_InceptionNet/' 


# build model
inceptionNet_model = InceptionNet(input_channle_size = 1,nb_classes = max(y_train)+1)



# build trainner
easy_use_trainner = Easy_use_trainner(Result_log_folder = Result_log_path,
                                      dataset_name = dataset_name,
                                      device = device)

# put model to trainner
easy_use_trainner.get_model(inceptionNet_model)


# fit data
easy_use_trainner.fit(X_train, y_train, X_test, y_test)


# test
y_predict = easy_use_trainner.predict(X_test)

print('correct:',y_test)
print('predict:',y_predict)
acc = accuracy_score(y_predict, y_test)
print(acc)
```
